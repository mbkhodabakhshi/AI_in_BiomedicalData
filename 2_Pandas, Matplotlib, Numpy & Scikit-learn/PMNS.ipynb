# Scikit

https://scikit-learn.org

## import

```python
import sklearn
sklearn.__version__
```

## dataset

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes

```python
from sklearn.datasets import load_diabetes

print(load_diabetes()['DESCR'])
```

```python
x, y= load_diabetes(return_X_y= True, as_frame= True)
x
```

```python
diabetes= load_diabetes()
print(diabetes.feature_names)
print(diabetes.target[0:20])
```

## split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target,
                                                    test_size= 0.2, random_state= 44)
x_train.shape, x_test.shape
```

##  regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
y_predict= lr.predict(x_test)
print(y_test[0:5])
print(y_predict[0:5].astype(int))
```

## metric

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_predict)
print(f'{mae:.2f}')
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_predict)
print(f'R²: {r2:.2f}')
```

-------------------------------------------------------------

## comparison of models

```python
from sklearn.linear_model import Ridge, Lasso, BayesianRidge, SGDRegressor, ElasticNet, PoissonRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_decomposition import PLSRegression
```

```python
def regresion_fn(model):
    print(model)
    model.fit(x_train, y_train)
    y_predict= model.predict(x_test)

    mae = mean_absolute_error(y_test, y_predict)
    print(f'MAE: {mae:.2f}')
    r2 = r2_score(y_test, y_predict)
    print(f'R²: {r2:.2f}')
    print('---------------------------')
```

```python
models= [LinearRegression(), Ridge(), Lasso(), SVR(), DecisionTreeRegressor(),
         RandomForestRegressor(), ElasticNet(), BayesianRidge(), PoissonRegressor(),
         SGDRegressor(), PLSRegression(n_components=2)]
for item in models:
    regresion_fn(item)
```

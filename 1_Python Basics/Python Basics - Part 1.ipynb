## 3.2 Comparison Operators

````python
print("3 >  2:",3 > 2)
print("3 >= 2:",3 >= 2)
print("3 <  2:",3 < 2)
print("3 <  2:",2 < 3)
print("3 <= 2:",2 <= 3)
print("3 == 2:",3 == 2)
print("3 != 2:",3 != 2)
````

````python
print('1 is 1:', 1 is 1)
print('1 is not 2:', 1 is not 2)
print('A in Alireza:', 'A' in 'Alireza')
print('R in Alireza:', 'R' in 'Alireza')
print('Is coding NOT in coding for all:','coding' not in 'coding for all')
print('4 is 2 ** 2:', 4 is 2 ** 2)
````

````python
print("3 > 2 and 4 > 3:",3 > 2 and 4 > 3)
print("3 > 2 and 4 < 3:",3 > 2 and 4 < 3)
print("3 < 2 and 4 < 3:",3 < 2 and 4 < 3)
print('True and True:  ', True and True)
print("3 > 2 or 4 > 3: ",3 > 2 or 4 > 3)
print("3 > 2 or 4 < 3: ",3 > 2 or 4 < 3)
print("3 < 2 or 4 < 3: ",3 < 2 or 4 < 3)
print('True or False:  ', True or False)
print("not 3 > 2:      ",not 3 > 2)
print("not True:       ",not True)
print("not False:      ",not False)
````

## 4.2 Function Returning a Value

````python
def generate_full_name():
    first_name = 'Alireza'
    last_name = 'Javadi'
    space = ' '
    full_name = first_name + space + last_name
    return full_name

name = generate_full_name()
````

````python
programming_lan = ['C', 'C++', 'Python', 'R']

print('Programming Languages:', programming_lan)
print(len(programming_lan))
print("First Programming Language in the list is:", programming_lan[0])

print(programming_lan[1:])
````
